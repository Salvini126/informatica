7.l’elenco dei consolati presenti in Argentina e in Brasile

8.i consolati di cui non si conosce il sito web

9.i consolati che non hanno una mail del Ministero degli Esteri

```python
result = db.consolati.find({"email":{"$not":{"$regex":"@esteri.it"}}},{"email":1})
df = pd.DataFrame(list(result))
df
```

10.il numero di consolati in ogni continente

```python
result=db.consolati.aggregate([  
                                   { "$group": { "_id": "$continente","total":{"$sum":1}}}, 
                                   {"$sort":{"_id":1}}
                                   ])
df = pd.DataFrame(list(result))
df
```

11.i continenti che hanno più di 50 consolati. Ordinare in ordine decrescente a partire dal continente che ha più consolati

```python
result=db.consolati.aggregate([  
                                   { "$group": { "_id": "$continente","total":{"$sum":1}}}, 
                                   {"$sort":{"total":-1}},
                                   {"$match":{"total":{"$gte":50}}},
                                   ])
df = pd.DataFrame(list(result))
df
```
