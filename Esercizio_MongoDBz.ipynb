Estrarre dalla collection le informazioni richieste:

```
# gli appartamenti che costano più di 100 $ a notte
```

```
df2 = pd.DataFrame(list(db.listingsAndReviews.find({"price": {'$gt' : 100 }})))
df2.head(1)
```

```
# gli appartamenti che costano tra i 100 $ e i 200 $ a notte
```

```
df3 = pd.DataFrame(list(db.listingsAndReviews.find({"price": {'$gt' : 100, '$lt' : 200}})))
df3.head(1)
```

```
# gli appartamenti che hanno almeno 5 posti letto e che costano tra i 50 $ e 100 $ a notte
```

```
df4 = pd.DataFrame(list(db.listingsAndReviews.find({"$and" : [{"price" : {'$gt' : 50, '$lt' : 100}}, {"beds" : {'$gt' : 5 }}]})))
df4.head(1)
```

```
# gli appartamenti in Portogallo
```

```
df5 = pd.DataFrame(list(db.listingsAndReviews.find({'address.country_code' : 'PT'})))
df5.head(1)
```

```
df6 = pd.DataFrame(list(db.listingsAndReviews.find({"address.country_code": {'$not' : {"$in": ["PT", "ES"]}}})))
df6.head(1)
```
